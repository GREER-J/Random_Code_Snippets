```python
#Imports
import random
import time
from matplotlib import pyplot as plt

#Variables
chance_of_ender_pearl = 20 / 423
chance_of_blaze = .5
amount_of_trades = 262
target_result = 41

continue_var = True
number_of_trials= 0
max_pearls = 0
cutoff = 60 * (60)
start_time = time.time()
time_of_last_report = start_time
data = []
max_pearls_data = []
number_of_sims_data = []

while(continue_var):
    #Check exit conditions
    run_start = time.time()
    if(run_start - start_time > cutoff):
        print("Emergency cutoff")
        continue_var = False
    
    if(run_start - time_of_last_report > 60):
        time_of_last_report = run_start
        print("Minute {}: Number of simulations {} | Max pearls {}".format(round((run_start - start_time)/60), number_of_trials, max_pearls))
    
    if(max_pearls >= target_result):
        print("Objective achieved!")
        continue_var = False
        
    
    
    #reset run variables
    number_of_trials += 1
    number_of_pearls = 0
    
    for i in range(amount_of_trades):
        random_num = random.random()
        if(random_num <= chance_of_ender_pearl):
            number_of_pearls += 1
    
    #Save data
    data.append(number_of_pearls)
    max_pearls_data.append(max_pearls)
    number_of_sims_data.append(number_of_trials)
    
    #Update max pearls
    if(number_of_pearls > max_pearls):
        max_pearls = number_of_pearls

        
plt.plot(data, label = "Amount of pearls")
plt.plot(max_pearls_data, label = "Max amount of pearls")
#plt.plot(number_of_sims_data, label = "Number of simulations")
plt.legend(loc = 'best')
plt.show()

plt.hist(data, density=True, bins=30)  # density=False would make counts
plt.ylabel('Probability')
plt.xlabel('Data');
plt.show()
```
