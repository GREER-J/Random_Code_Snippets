Task:
The town of Shelbyville has a population of 100,000. The town already has a wind farm which generates electricity proportional to wind speed. The town also has a dam producing a set amount of electricity. The town wish to store any excess power in a holding tank in the local mountains. This way they are able to recapture that energy and use it later.

How big does the town have to be before the chance of a power outage in 100 years is below 1%?

```python
import numpy as np
import random


#Variables
average_wind = 11
standard_dev_wind = 8
state = "windy"
res_cap_min = 100000
res_cap_max = 1000000

#tracked variables
blackout_hours = 0
years = 1

def get_power_demand(hour):
    demand = 0
    if(hour >= 6 and hour <= 8):
        demand = 1000
    if(hour >= 9 and hour <= 11):
        demand = 400
    if(hour >= 12 and hour <= 16):
        demand = 500
    if(hour >= 17 and hour <= 21):
        demand = 1000
    return demand


def get_wind_speed(average_wind, standard_dev_wind):
    wind_velocity = round(np.random.normal(average_wind, standard_dev_wind, 1)[0],0)
    if(wind_velocity < 0):
        wind_velocity = 0
    return(wind_velocity)


def get_wind_power(wind_velocity):
    wind_power = 0
    if(wind_velocity <= 4 or wind_velocity > 25):
        wind_power = 0
    if(wind_velocity > 16 and wind_velocity <= 25):
        wind_power = 2 * 2100
    else:
        wind_power = (2100 * (wind_velocity - 4) / 12) * 2
    if(wind_power < 0):
        wind_power = 0
    return(wind_power)


def set_state(state):
    rn = random.random()
    if(state == "windy"):
        if(rn > 0.8):
            state = "still"
    else:
        if(rn > 0.9):
            state = "windy"
    return(state)

res_cap = 0.75 * res_cap_max
dam_level = 0.5 * res_cap

for day in range(365 * years):
    for hour in range(23):
        state = set_state(state)
        if(state == "windy"):
            wind_velocity = get_wind_speed(average_wind, standard_dev_wind)
            wind_power = get_wind_power(wind_velocity)
            actual_power = wind_power
        else:
            actual_power = 0
        power_demand = get_power_demand(hour)
        net_power = actual_power - power_demand

        if(net_power == 0):
            #print("Perfectly balanced as all things should be")
            pass
        if(net_power > 0):
            if(dam_level < res_cap):
                #Pump water up to resiviour
                dam_level += net_power
                #print("Pump water up to dam")
            else:
                print("Dam full!")
        if(net_power < 0):
            if(dam_level <= 0):
                #print("Blackout!")
                blackout_hours += 1
            if(dam_level > 0):
                #Calc how much water needed to be let out of dam
                dam_level -= net_power
                #print("Pump water from dam")
        print(hour, dam_level)
        

```
