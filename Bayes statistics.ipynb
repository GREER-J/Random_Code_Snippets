```python
#Imports
from matplotlib import pyplot as plt

#Variables
percent_change = []
prior_odds = []

#Functions
def baysian(p_hypothosis: float, p_hypothosis_given_event: float, p_not_hypothosis_given_event: float):
    num = (p_hypothosis * p_hypothosis_given_event)
    den = (p_hypothosis * p_hypothosis_given_event + (1 - p_hypothosis) * p_not_hypothosis_given_event)
    updated_prob = round(num / den, 3)
    return(updated_prob)

#Calculation
for p in range(0, 101):
    delta = baysian(p/100, 0.95, 0.25) - p/100
    percent_change.append(delta)
    prior_odds.append(p/100)

#Ouptut
plt.plot(prior_odds, percent_change)
plt.show()
```
